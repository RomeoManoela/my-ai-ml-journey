```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, binom, poisson, norm, uniform, lognorm
```

```python
# Bernoulli distribution
sns.set_style("darkgrid")
x = [0, 1]
probs = bernoulli.pmf(x, p=0.4)
plt.bar(x, probs)
plt.title("Bernoulli Distribution")
plt.xlabel("x")
plt.ylabel("P(X = x)")
plt.show()
```

```python
# Binomial distribution
n = 10         # nombre d'essais
p = 1/6        # probabilité de succès (obtenir un 6)
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y)
plt.title("Binomial Distribution: 10 lancers, succès=6 (p=1/6)")
plt.xlabel("Nombre de fois où le 6 est obtenu")
plt.ylabel("Probabilité")
sns.set_style("darkgrid")
plt.show()
```

```python
# Poisson distribution
lambda_ = 5
x = np.arange(0, 15)
y = poisson.pmf(x, lambda_)

plt.bar(x, y)
plt.title("Poisson Distribution: lambda=5")
plt.xlabel("Nombre d'occurrences")
plt.ylabel("Probabilité")
sns.set_style("darkgrid")
plt.show()
```

```python
# Normal Distribution
mu = 1.7
sigma = 0.3
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 50)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.title("Normal Distribution: mu=1.7, sigma=0.3")
plt.xlabel("x")
plt.ylabel("P(X = x)")
sns.set_style("darkgrid")
plt.fill_between(x, y, alpha=0.3)
plt.show()
```

```python
# Uniform Distribution
a = 0
b = 1
x = np.linspace(a, b, 50)
y = uniform.pdf(x, a, b)

plt.plot(x, y)
plt.title("Uniform Distribution: a=0, b=1")
plt.xlabel("x")
plt.ylabel("P(X = x)")
sns.set_style("darkgrid")
plt.fill_between(x, y, alpha=0.3)
plt.show()
```

```python
# log normal
mu = 0
sigma = 1
x = np.linspace(0, 5, 50)
y = lognorm.pdf(x, sigma, scale=np.exp(mu))

plt.plot(x, y)
plt.title("Log Normal Distribution: mu=0, sigma=1")
plt.xlabel("x")
plt.ylabel("P(X = x)")
sns.set_style("darkgrid")
plt.fill_between(x, y, alpha=0.3)
plt.show()
```
